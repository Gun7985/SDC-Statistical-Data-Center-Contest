시도코드 -> 시도명 컬럼 만들기

```python
# 코드와 시도명을 매핑하는 딕셔너리 생성
code_to_name = {
    "00": "전국",
    "11": "서울특별시",
    "21": "부산광역시",
    "22": "대구광역시",
    "23": "인천광역시",
    "24": "광주광역시",
    "25": "대전광역시",
    "26": "울산광역시",
    "29": "세종특별자치시",
    "31": "경기도",
    "32": "강원도",
    "33": "충청북도",
    "34": "충청남도",
    "35": "전라북도",
    "36": "전라남도",
    "37": "경상북도",
    "38": "경상남도",
    "39": "제주특별자치도"
}

# 행정구역시도코드 컬럼을 문자열로 변환한 후 매핑 적용
df1['행정구역시도명'] = df1['행정구역시도코드'].astype(str).map(code_to_name)
df2['행정구역시도명'] = df2['행정구역시도코드'].astype(str).map(code_to_name)
```

```python
import pandas as pd

# 중복되는 시군구 추출
duplicated_sigungus = code[code.duplicated(subset=['시군구'], keep=False)]

# 중복된 시군구의 전체 데이터 확인
duplicated_sigungus
```

```python
# df1의 시도코드와 시군구코드를 결합하여 5자리 문자열로 만듦
df1['시군구코드'] = (df1['행정구역시도코드'].astype(str).str.zfill(2) +
                       df1['행정구역시군구코드'].astype(str).str.zfill(3))

# `code` 데이터프레임의 시군구코드를 문자열로 변환
code['시군구코드'] = code['시군구코드'].astype(str)

# `df1`와 `code`를 'combined_code'와 '시군구코드'를 기준으로 병합
merged_df1 = df1.merge(code, left_on='시군구코드', right_on='시군구코드', how='left')

# 필요한 컬럼 선택 및 이름 변경
merged_df1 = merged_df1[['가구일련번호','행정구역시도코드', '행정구역시군구코드', '시군구코드', '성별코드',
                         '만연령','행정구역시도명','시도_시군구']]
merged_df1.rename(columns={'시도_시군구': '행정구역시군구명'}, inplace=True)

# 결과 출력
merged_df1
```
